```python
import pandas as pd
```

```python
data_f=pd.read_excel(r"C:\Users\hp\Downloads\sad.xlsx")
data_f
```

```python
cdf = data_f[['Birth','Population_2018']]
new_data=cdf.head(50)
```

```python
new_data=cdf.head(55)
df = new_data[['Birth','Population_2018']]
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.plot(df.Population_2018, df.Birth,  color='blue')
plt.scatter(df.Population_2018, df.Birth,  color='blue')
plt.xlabel("Population")
plt.ylabel("Birth")
plt.show()
```

```python
import numpy as np
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(df[['Population_2018']])
train_y = np.asanyarray(df[['Birth']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(df.Population_2018, df.Birth,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Population")
plt.ylabel("Birth")
```

```python
msk = np.random.rand(len(new_data)) < 0.8
train = df[msk]
test = df[~msk]
print(train)
print(test)
```

```python
import numpy as np
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Population_2018']])
train_y = np.asanyarray(train[['Birth']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.Population_2018, train.Birth,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Population")
plt.ylabel("Birth")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['Population_2018']])
test_y = np.asanyarray(test[['Birth']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))# original test set output vs predicted test set output using tuned parameters
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
