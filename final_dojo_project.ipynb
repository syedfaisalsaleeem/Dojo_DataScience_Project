```python
import pandas as pd
data_f=pd.read_excel(r"C:\Users\hp\Downloads\dojo.xlsx")
data_f
```

```python
cdf = data_f[['Birth_n','Population_2018']]
new_data=cdf.head(45)
```

```python
new_data=cdf.head(45)
df = new_data[['Birth_n','Population_2018']]
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.plot(df.Population_2018, df.Birth_n,  color='blue')
plt.scatter(df.Population_2018, df.Birth_n,  color='blue')
plt.xlabel("Population")
plt.ylabel("Birth")
plt.show()
```

```python
import numpy as np
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(df[['Population_2018']])
train_y = np.asanyarray(df[['Birth_n']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(df.Population_2018, df.Birth_n,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Population")
plt.ylabel("Birth")
```

```python
test_y_hat = regr.predict([[0.6*10**7]])
print(test_y_hat)
```
